```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
import datetime
from datetime import date
```

```python
movie_data = "Movie_Last_Clean.csv"
movie_data_df= pd.read_csv(movie_data)
movie_data_df.head()
```

```python
movie_data_df.dropna(axis=0,how='any')
#movie_data_df.count
```

```python
ax1=yr2019_df.groupby(["Studio"],as_index=False).agg({'Opening Gross Amount':'sum'}).plot(kind='line',figsize = (18,10))
ax1.grid()
plt.title("Opening Gross Amount throughout 2019")
plt.xlabel("Section in Year")
plt.ylabel("Gross Amount")

plt.savefig("2019_Gross.png")
#plt.show()
```
