```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
import datetime
from datetime import date
```

```python
movie_data = "Movie Data Cleaned.csv"
movie_data_df= pd.read_csv(movie_data)
#movie_data_df.head()
```

```python
movie_data_df["Runtime"]=movie_data_df["Runtime"].str.rstrip(" min")
movie_data_df=movie_data_df.rename(columns={"Runtime":'Runtime In Min'})
```

```python
movie_data_df["Rank"]=movie_data_df["Rank"].astype(int)
movie_data_df["Open"]=movie_data_df["Open"].replace({'\$': '',',': ''}, regex=True)
movie_data_df["Total Gross"]=movie_data_df["Total Gross"].replace({'\$': '',',': ''}, regex=True)
```

```python
movie_data_df=movie_data_df.rename(columns={"Open":"Opening Day Date"})
movie_data_df["Total Gross"]=movie_data_df["Total Gross"].astype(float)
movie_data_df=movie_data_df.rename(columns={"O Theaters":"Amount of Theaters on Opening Day"})
movie_data_df=movie_data_df.rename(columns={"Opening":"Opening Gross Amount"})
movie_data_df["Opening Gross Amount"]=movie_data_df["Opening Gross Amount"].replace({'\$': '',',': ''}, regex=True)
movie_data_df["Opening Gross Amount"]=movie_data_df["Opening Gross Amount"].astype(float)
movie_data_df["imdbVotes"]=movie_data_df["imdbVotes"].replace({'\$': '',',': ''}, regex=True)
movie_data_df["imdbVotes"]=movie_data_df["imdbVotes"].astype(float)
movie_data_df=movie_data_df.rename(columns={"TG Theaters":"Total Theaters"})
```

```python
movie_data_df["Runtime In Min"]=movie_data_df["Runtime In Min"].astype(float)
```
